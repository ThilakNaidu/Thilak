## Pre-processing and Creating pipeline

```python
# Importing Regex Package
import regex as re
# Defining a Function to Pull Words
def tokenize(CONTENT):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', CONTENT)
```

```python
# Importing nltk Package
import nltk
```

```python
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
```

```python
# Creating a Pipeline
pipeline = [str.lower, tokenize, remove_stop]

# Defining a Function to Run the Pipeline
def prepare(CONTENT, pipeline):
    if isinstance(CONTENT, str):  # Check if description is a string
        tokens = CONTENT
        for fun in pipeline:
            tokens = fun(tokens)
        return tokens
    else:
        return []
```

```python
# Running the Pipeline
df['tokens'] = df['CONTENT'].apply(prepare, pipeline=pipeline)
```

```python
#creating a New Column for the Number of Tokens
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

## Cleaning the Data

```python
#Clean the text

#Create a function to clean the content
def cleanTxt(CONTENT):
    text = re.sub(r'@[A-Za-z0-9]+', '',CONTENT ) #Remove @mentions replace with blank
    text = re.sub(r'#', '', CONTENT) #Remove the '#' symbol, replace with blank
    text = re.sub(r'RT[\s]+', '', CONTENT) #Removing RT, replace with blank
    text = re.sub(r'https?:\/\/\S+', '', CONTENT) #Remove the hyperlinks
    text = re.sub(r':', '', CONTENT) # Remove :
    
    return CONTENT

#Cleaning the content
df["CONTENT"]= df["CONTENT"].apply(cleanTxt)

#Show the clean content
df.head()
```

## Sentiment Analysis on Dataframe

```python
from textblob import TextBlob
from wordcloud import WordCloud, STOPWORDS
```

```python
#Create a function to get the subjectivity
def getSubjectivity(CONTENT):
    return TextBlob(CONTENT).sentiment.subjectivity

#Create a function to get Polarity
def getPolarity(CONTENT):
    return TextBlob(CONTENT).sentiment.polarity

#Now we create a new column for what we just did and add it to the df dataframe
df['Subjectivity'] = df['CONTENT'].apply(getSubjectivity)
df['Polarity'] = df['CONTENT'].apply(getPolarity)

#Now display data
df.head()
```

```python
#Group the range of Polarity to different categories
def getInsight(score):
    if score < 0: # If the polarity score is less than 0, "Negative"
        return "Negative"
    elif score == 0: # If the polarity score is Equal to 0, "Equal"
        return "Neutral"
    else: # If the polarity score is more than to 0, "Positive"
        return "Positive"
# Applying the getInsight function to the 'Polarity' column and creating a new column 'Insight' to store   
df["Insight"] = df["Polarity"].apply(getInsight)

# Displaying the first 50 rows
df.head(50)
```

## Creation of Dictionary for Sentiment Analysis

```python
# Let's create a dictionary which we can use for scoring our review content

# Downloading punkt from NLTK library
nltk.download('punkt')

# Renaming the column 'reviewcontent' to 'CONTENT' in the DataFrame
df.rename(columns={"reviewcontent": "CONTENT"}, inplace=True)

# Assigning positive and negative scores
pos_score = 1
neg_score = -1

# Initializing an empty dictionary
word_dict = {}
 
# Adding the positive words to the dictionary
for word in opinion_lexicon.positive():
        word_dict[word] = pos_score
      
# Adding the negative words to the dictionary
for word in opinion_lexicon.negative():
        word_dict[word] = neg_score
```

## Calculating Sentiment Score with Bing Liu Lexicon

```python
#Creating a fuunction content
def bing_liu_score(CONTENT):
    #Initializing the sentiment score
    sentiment_score = 0
    #Tokenizing the input conetnt into words and convert them to lowercase
    bag_of_words = word_tokenize(CONTENT.lower())
     # creating loop to check each word in the bag of words
    for word in bag_of_words:
        #Checking if the word exists in the sentiment dictionary
        if word in word_dict:
            # If the word exist, adding its sentiment score to the sentiment score
            sentiment_score += word_dict[word]
    return sentiment_score  #Returning the sentiment score for the content
```

```python
# Fill NaN values in the 'content' column
df['CONTENT'].fillna('no review', inplace=True)
#creating new column 'Bing_Liu_Score' to store the scores by applying  bing_liu_score to calculate sentiment scores for each text
df['Bing_Liu_Score'] = df['CONTENT'].apply(bing_liu_score)
```

## Calculating Mean Sentiment Score

```python
# Grouping 'Class' rating and calculating the mean sentiment score using the Bing Liu Lexicon
df.groupby('CLASS').agg({'Bing_Liu_Score':'mean'})
```

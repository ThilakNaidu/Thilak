## Pre-processing pipeline

```python
# Importing Regex Package
import regex as re
# Defining a Function to Pull Words
def tokenize(description):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', description)
```

```python
import nltk
```

```python
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
```

```python
# Creating a Pipeline
pipeline = [str.lower, tokenize, remove_stop]

# Defining a Function to Run the Pipeline
def prepare(description, pipeline):
    if isinstance(description, str):  # Check if description is a string
        tokens = description
        for fun in pipeline:
            tokens = fun(tokens)
        return tokens
    else:
        return []  # Return an empty list for non-string values
```

```python
# Running the Pipeline
df['tokens'] = df['description'].apply(prepare, pipeline=pipeline)
```

```python
#creating a New Column for the Number of Tokens
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

## Word Frequency Analysis¶

```python
# importing a function Counter
from collections import Counter
```

```python
# Printing the 5 Words Repeated the Most
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
# Creating a New Data Frame for Frequency Analysis 
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

```python
# Sorting the freq_df based on Frequency
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
# Printing the Statistical Data of Frequency
freq_df.describe()
```

```python
# Plotting a bar graph for most frequent words 
ax = freq_df.head(10).plot(kind='barh', width=0.95, figsize=(8,6))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```
